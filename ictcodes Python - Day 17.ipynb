# Task 1: Change Password

Enter Current Password: ******

if Current Password is True

Then

Enter New Password:

Re-Enter New Password:




```
pf=open("Password.txt","r")
password = pf.read()
pf.close()

p = input("Enter Password: ")
if(p==password):
  x=input("Enter New Password: ")
  y=input("Confirm Password: ")

  if(x==y):
    g=open("Password.txt","w")
    g.write(x)
    g.close()
    print("Password Changed!!!")
  else:
    print("Passwords Does not Match!")

else:
  print("Wrong Passowrd!!")


```

# Task 2 : Fix the Single Digit Issue

```
# File Writing Part
myfile = open("MyData.txt","w")

for i in range(10):
  k=int(input("Enter Marks: "))
  if(k<10):
    k = "0" + str(k)
  myfile.write(str(k))

myfile.close()


```

# NumPy

NumPy is a python library for working with Arrays.







```
import numpy

#Zero Dimensional Array
arr0d = numpy.array(59)

#One Dimensional Array
arr1d = numpy.array([1,2,3,4,5,6])

#Two Dimensional Array
arr2d = numpy.array([[1,2,3] ,[4,5,6]])

#Three Dimensional Array
arr3d = numpy.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(arr0d)
print(arr1d)
print(arr2d)
print(arr3d)

print(arr0d.ndim)
print(arr1d.ndim)
print(arr2d.ndim)
print(arr3d.ndim)

```
