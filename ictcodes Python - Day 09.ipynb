# Date Module

https://www.w3schools.com/python/python_datetime.asp

```
import datetime as dt

x = dt.datetime.now()
print(x)
print(x.strftime("%M"))

a = dt.datetime(1998,2,3)
print(a.strftime("%A"))


```

```
from datetime import date

a = date(1995,10,4)
b = date(2020,10,10)

d = a-b
print(d.days)

```
