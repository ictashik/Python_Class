#Task: Sorting a List 

There are 10 Numbers in a list, 

Rearrange the Numbers in Ascending Order


```
#Answer Code:

# Bubble Sort
# Merge Sort
# Shell Sort
# Selection Sort
# Insertion Sort


def bubblesort(list):

    for i in range(len(list)-1,0,-1):
        for j in range(i):
            if list[j]>list[j+1]:
                temp = list[j]
                list[j] = list[j+1]
                list[j+1] = temp


list = [19,2,31,45,6,11,121,27]
bubblesort(list)
print(list)
```

#NumPy Sorting

```
import numpy as np

arr = np.array([36,22,14,5,3,88,66,105,194])
#Ascending Order
print(np.sort(arr))

#Descending Order
print(np.sort(arr)[::-1])
```
