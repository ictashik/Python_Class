# Opertators

Operators are the main building block of any programming language. Operators allow the programmer to perform different kinds of operations on operands. These operators can be categorized based upon their different functionality.

### Arithmetic operators:
Arithmetic operators are used to perform mathematical operations like addition, subtraction, multiplication and division.

```python
a = 9
b = 4
   

add = a + b 
sub = a - b 
mul = a * b 
divfloat = a / b 
divfloor = a // b 

mod = a % b 
   
# print results 
print(add) 
print(sub) 
print(mul) 
print(divfloat) 
print(divfloor) 
print(mod)
```

##Relational Operators:
 Relational operators compares the values. It either returns True or False according to the condition.

```python
a = 14
b = 28
   
print(a > b) 
   
print(a < b) 
   
print(a == b) 
   
print(a != b) 
   
print(a >= b)    

print(a <= b)
```

## Logical Operators: 
Logical operators perform Logical AND, Logical OR and Logical NOT operations.

```python
a = True
b = False
   
print(a and b) 
   
print(a or b) 
   
print(not a) 
```

##Assignment operators:

##Special operators: 

Special operators are of two types-
Identity operator that contains is and is not.
Membership operator that contains in and not in.

##Bitwise operators:

Bitwise operator acts on bits and performs bit by bit operation.

# Build A Simple Interest Calculator

Now Let us build a Simple Interest calculator in python

**Interest = Principal  x No. of Years x Rate (%)**


```python
p=float(input("Enter Principal Amount: "))
n=float(input("Enter No of Years: "))
r=float(input("Enter Rate % : "))

interest  = p*n*r
total = p + interest
   
print("Interest = ", interest)
print("After ",n, " years you will recieve ", total)
```

# Build A Compund Interest Calculator

Now Let us build a Compound Interest calculator in python

Compound Interest = Principal  x [1 + Rate(%)] ^ No.of.Years 

*Use pow(a,b) to compute a^b*


```python
p=float(input("Enter Principal Amount: "))
n=float(input("Enter No of Years: "))
r=float(input("Enter Rate % : "))

interest  = p*(pow(1+r,n))
total = p + interest
   
print("Interest = ", interest)
print("After ",n, " years you will recieve ", total)
```
