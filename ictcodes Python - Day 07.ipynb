# Task: Recode Mark Calculator by inputting Number of Subject

Enter Name:

How many Subjects:

Same as above

```
#answer code
name=input("Enter your Name: ")
n=int(input("Enter Number of Subjects: "))
m=int(input("Enter Maximum Marks in one Subject: "))

total=0
mark = []

for x in range(n):
  p=float(input("Enter Marks: "))
  mark.append(p)
  total=total+mark[x]


average = total/n
percentage=(total/(m*n))*100

print("\n\nHello ",name)
print("Your Marks are ",mark)
print("\n\nTotal Mark = ", total)
print("Average Mark = ",average)
print("Percentage = ",percentage,"%")
```

# Recursive Functions


```
def area(l,b):
  c=l*b
  return c

def vol(a,h):
  d=a*h
  return d

print(vol(area(15,20),65))
```

# Factorial using Recursive Functions


```
def factorial(x):
  if x==1:
    return 1
  else:
    return (x * factorial(x-1))
  
k=int(input("Enter a Number: "))
print(factorial(k))


```

#Lambda Functions


func =  lambda agruments : expressions


```
#Normal Function
def xt(b):
  c=b+20
  return c

#Lambda Function
x = lambda a : a+20

print(x(10))
print(xt(10))
```

```
#lambda function using multiple parameters/arguments

mul = lambda a,b : a*b

print(mul(20,30))

```
