```
pf=open("password.txt","r")
password = pf.read()
pf.close()

p=input(" Enter Password to Continue: ")

if(p==password):
  k = int(input("Enter a Number "))
  for i in range(10):
    x = str(i+1) + " x " + str(k)  + str(" = ") + str((i+1)*k)
    print(x)
else:
  print("Wrong Password!!!")
```

# Python Turtle Graphics


```
import turtle
t = turtle.Turtle()

#Forward  -- t.fd
#Backward  -- t.bk
#Left  -- t.lt
#Right  -- t.rt

t.pensize(5)

t.forward(20)
t.left(90)
t.forward(30)
t.left(30)
t.forward(30)
t.right(100)
t.backward(40)

t.lt(50)
t.fd(150)
t.rt(300)
t.bk(300)
```
